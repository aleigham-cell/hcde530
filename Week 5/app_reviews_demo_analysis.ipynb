# App Reviews CSV Analysis
This notebook inspects `app_reviews_demo.csv` and answers the requested questions about the dataset structure, distribution, filtered subset, grouped averages, and missing values.

```python
import pandas as pd

#Load the data
path = 'app_reviews_demo.csv'
df = pd.read_csv(path)
```

## Inspect the Dataset Structure
Use `head()` to preview the first rows and `info()` to see column types and non-null counts.

```python
print("Dataset shape:", df.shape)
#Data contains data about app reviews, including the app name, category, rating, review text, and helpful votes
df.head()
```

```python
df.info()
#data contains a mixture of numeric and text data, with some missing values. The 'Rating' column is a float, while the 'HelpfulVotes' column is an integer
```

## Analyze the Distribution of the Most Important Column
We treat `rating` as the most important column for review sentiment and calculate its distribution.

```python
# Analyze the distribution of ratings (most important column)
rating_distribution = df['rating'].value_counts().sort_index()
print(rating_distribution)
#Most common rating is 5, followed by 4
```

## Filter a Meaningful Subset
Filter reviews for a meaningful subset, such as high-rated field research reviews.

```python
# Filter the dataset for 'field research' category and ratings of 4 or higher
subset = df[(df['category'] == 'field research') & (df['rating'] >= 4)]
print("Subset shape:", subset.shape)
print(subset.head())
```

## Group by Category and Compute Numeric Averages
Group by `category` and compute the average of `helpful_votes` for each category.

```python
# Calculate the average rating for each app
app_rating_mean = df.groupby('app')['rating'].mean().round(2)
print(app_rating_mean)
```

## Locate Missing Values and Incomplete Columns
Check which columns have missing values and whether any columns are incomplete.

```python
missing_counts = df.isna().sum()
print(missing_counts)
print()
print("Columns with missing values:")
print(missing_counts[missing_counts > 0])
```
